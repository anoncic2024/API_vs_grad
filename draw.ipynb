```python
from pathlib import Path
import re

accuracy_runs = []
primes = []
for nn in range(1,21):
    file = "./30_clip_nonlip/{}.txt".format(nn)
    data = Path(file).read_text()
    lst = re.findall(r"val_acc: (0\.\d+)", data)
    

    acc = []
    for h in lst:
        acc.append(float(h))
    accuracy_runs.append(acc)
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

# Plotting training and testing losses
plt.figure(figsize=(8, 5))
colors = cm.plasma(np.linspace(0, 1, len(range(1,21))))
# plot one horizontal line as 1/12 dashed line
#plt.axhline(y=1/12, color='r', linestyle='--', label='baseline 1/12')  
plt.legend(title = "Legend Title") 
plt.title(r"bitsize=30, non-lipschitz case, 20 runs", fontsize=18) 
for i, A in enumerate(range(1,21)):
    plt.plot(np.linspace(0, 1000, 100), accuracy_runs[i], linewidth=0.7, color=colors[i])
plt.xlabel('Epoch', fontsize=18)
plt.ylabel('accuracy', fontsize=18)
# save plot
plt.ylim(0.4, 1.0)
plt.legend()
plt.savefig('figs/picture.pdf', format="pdf", dpi=400, bbox_inches='tight')
plt.show()
```
